```python
class Node:
    def __init__(self, u):
        self.data = u
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

    def create(self, root, x):
        if root is None:
            return Node(x)
        elif root.data > x:
            root.left = self.create(root.left, x)
        else:
            root.right = self.create(root.right, x)
        return root

    def inorder(self, root):
        if root:
            self.inorder(root.left)
            print(root.data, end=' ')
            self.inorder(root.right)

    def preorder(self, root):
        if root:
            print(root.data, end=' ')
            self.preorder(root.left)
            self.preorder(root.right)

    def postorder(self, root):
        if root:
            self.postorder(root.left)
            self.postorder(root.right)
            print(root.data, end=' ')
            

    def sum_(self, root):
        if root is None:
            return 0
        return root.data + self.sum_(root.left) + self.sum_(root.right)
        

    def sum_even(self, root):
        if root is None:
            return 0
        total = 0
        if root.data % 2 == 0:
            total += root.data
        total += self.sum_even(root.left)
        total += self.sum_even(root.right)
        return total
    
    def diff_sum(self,root):
        if root==None:
            return 0
        if root.data%2==0:
            return root.data+self.diff_sum(root.left)+self.diff_sum(root.right)
        return self.diff_sum(root.left)+self.diff_sum(root.right)-root.data
    
    def height(self,root):
        if root==None:
            return -1
        return max(self.height(root.left),self.height(root.right))+1
    
    def depth(self, root, x, current_depth=0):
        if root is None:
            return -1
        if root.data == x:
            return current_depth
        if x < root.data:
            return self.depth(root.left, x, current_depth + 1)
        return self.depth(root.right, x, current_depth + 1)

   
    def bal(self,root):
        return abs(self.height(root.left)-self.height(root.right)) <=1
    
    
    def count_leaf(self,root):
        if root==None:
            return 0
        if root.left==None and root.right==None:
            return 1
        return self.count_leaf(root.left)+self.count_leaf(root.right)
    
    
    def search(self, root, x):
        if root is None or root.data == x:
            return root
        if x < root.data:
            return self.search(root.left, x)
        return self.search(root.right, x)
        
        
    def left_view(self,root,c,li):
        if root==None:
            return 
        if c not in li:
            print(root.data)
            li.append(c)
        self.left_view(root.left,c+1,li)
        self.left_view(root.right,c+1,li)
        
        
        
    def right_view(self,root,c,li):
        if root==None:
            return 
        if c not in li:
            print(root.data)
            li.append(c)
        self.right_view(root.right,c+1,li)
        self.right_view(root.left,c+1,li)   
        
        
    def top_view(self,root,c,dict1):
        if root==None:
            return 
        if c not in dict1:
            print(root.data)
            dict1.update({c:root.data})
        self.top_view(root.left,c+1,dict1)  
        self.top_view(root.right,c-1,dict1) 
    

t1 = Tree()
t1.root = t1.create(t1.root, 10)
t1.root = t1.create(t1.root, 5)
t1.root = t1.create(t1.root, 15)
t1.root = t1.create(t1.root, 3)
t1.root = t1.create(t1.root, 7)
t1.root = t1.create(t1.root, 12)
t1.root = t1.create(t1.root, 18)


li=[]
print("left view:")
t1.left_view(t1.root,0,li)


li=[]
print("right view:")
t1.right_view(t1.root,0,li)


print()
dict1={}
t1.top_view(t1.root,0,dict1)
print("Preorder traversal: ")
t1.preorder(t1.root)
print("\nInorder traversal: ")
t1.inorder(t1.root)
print("\nPostorder traversal: ")
t1.postorder(t1.root)
print()
print("difference between left and right subtrees:",abs((t1.sum_(t1.root.left))-(t1.sum_(t1.root.right))))

print("\nSum of all nodes: ", t1.sum_(t1.root))
print("Sum of even nodes: ", t1.sum_even(t1.root))

print("difference of sum:",t1.diff_sum(t1.root))
print("height od tree is:",t1.height(t1.root))
if t1.bal(t1.root):
    print("bal")
else:
    print("not balance")

print(t1.count_leaf(t1.root))
s=0

if t1.search(t1.root,5):
     print("true")
else:
     print("else")
```
