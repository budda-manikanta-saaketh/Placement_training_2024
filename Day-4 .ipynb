```python
class Node:
    def __init__(self, u):
        self.data = u
        self.nxt = None

class SLL:
    def __init__(self):
        self.head = None
    
    def display(self):
        t = self.head
        while t != None:
            print(t.data, end=" -> ")
            t = t.nxt
        print("")
    
    def add_back(self, x):
        if self.head is None:
            self.head = Node(x)
            return
        t = self.head
        while t.nxt != None:
            t = t.nxt
        t.nxt = Node(x)
    
    
    def add_even(self):
        t = self.head
        s = 0
        while t!=None:
            if t.data % 2 == 0:
                s += t.data
            t = t.nxt
        print("Sum of even numbers:", s)
        
    
    def search(self,k):
        t=self.head
        while t!=None:
            if t.data==k:
                return True
            t=t.nxt
        return False
    
    def addfront(self,x):
        t=Node(x)
        t.nxt=self.head
        self.head=t
        
        
    def mid_node(self):
        s=f=self.head
        while f and f.nxt:
            s=s.nxt
            f=f.nxt.nxt 
        return s.data
    
    
    def len_li(self):
        f,s=self.head,self.head
        while(f!=None and f.nxt!=None):
            s=s.nxt
            f=f.nxt.nxt
        if f==None:
            return("even")
        else:
            return("odd")
        
        
    def check(self):
        m=0
        c=0
        t=self.head
        while t.nxt!=None:
            if(t.nxt.data - t.data == 1):
                c+=1
            else:
                if c>m:
                    m=c
                c=1
            t=t.nxt
        if c>m:
            m=c
        return m
    
    
    def pairs(self):
        t=self.head
        while t !=None:
            s=t.nxt
            while s !=None:
                print(t.data,s.data)
                s=s.nxt
            t=t.nxt
        
    
    def bub_sort(self):
        T=self.head
        p=None
        c=0
        while(T!=p):
            f=0
            t=self.head.nxt
            while(t.nxt!=None):
                if t.data>t.nxt.data:
                    f=1
                    t.data,t.nxt.data = t.nxt.data,t.data
                t=t.nxt
                c+=1
            if f==0:
                break
            p=t
            T=T.nxt
            return c
        
                
            
            
l1 = SLL()
l1.add_back(1)
l1.add_back(2)
l1.add_back(3)
l1.add_back(5)
l1.add_back(6)

l2 = SLL()
l2.add_back(100)
l2.add_back(200)
l2.add_back(300)
l2.add_back(400)
l2.add_back(500)
l2.add_back(600)

print("Linked list 1:")
l1.display()
print("Linked list 2:")
l2.display()

l1.add_even()
l2.add_even()

print(l1.addfront(0))

print("Linked list 1:")
l1.display()

print(l1.search(10))

print(l1.mid_node())
print(l2.mid_node())

print(l1.len_li())

print(l1.check())

l1.pairs()

l1.bub_sort()

print("Linked list 1:")
l1.display()
```
